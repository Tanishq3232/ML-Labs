```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('car_evaluation.csv', names = ['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety', 'outcome'])
df.head()
```

```python
df.shape
```

```python
df.outcome.value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
en = LabelEncoder()
df.loc[:, :] = df.loc[:, :].apply(en.fit_transform)
df.head()
```

```python
X = df.iloc[:,:-1]
y = df['outcome']
X.head()
```

```python
from collections import Counter
Counter(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 10)

X_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
model.fit(X_train, y_train)
pred = model.predict(X_test)
```

```python
Counter(y_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, pred))
pd.crosstab(y_test, pred)
```

```python
import imblearn
from imblearn.under_sampling import RandomUnderSampler
under = RandomUnderSampler(sampling_strategy = {0: 60, 1: 60})
X_under, y_under = under.fit_resample(X, y)
```

```python
from sklearn.model_selection import train_test_split
X_under_train, X_under_test, y_under_train, y_under_test = train_test_split(X_under, y_under, test_size = 0.3, random_state = 10)
```

```python
X_under_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
model.fit(X_under_train, y_under_train)
pred_under = model.predict(X_under_test)
Counter(y_under_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_under_test, pred_under))
pd.crosstab(y_under_test, pred_under)
```

Over-sampling

```python
import imblearn
from imblearn.over_sampling import RandomOverSampler
over = RandomOverSampler(sampling_strategy = {0: 1210, 1: 1210, 3:1210})
X_over, y_over = over.fit_resample(X, y)
from collections import Counter
Counter(y_over)
```

```python
from sklearn.model_selection import train_test_split
X_over_train, X_over_test, y_over_train, y_over_test = train_test_split(X_over, y_over, test_size = 0.3, random_state = 10)
X_over_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
model.fit(X_over_train, y_over_train)
pred_over = model.predict(X_over_test)
Counter(y_over_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_over_test, pred_over))
pd.crosstab(y_over_test, pred_over)
```

SMOTE

```python
data=pd.read_csv('car_evaluation.csv')
data.head()
```

```python
X=data.iloc[:,:-1]
y=data.outcome
X.head()
```

```python
from sklearn.preprocessing import LabelEncoder
enc=LabelEncoder()

X['buying']=enc.fit_transform(X['buying'])
X['maint']=enc.fit_transform(X['maint'])
X['lug_boot']=enc.fit_transform(X['lug_boot'])
X['safety']=enc.fit_transform(X['safety'])
X['doors']=enc.fit_transform(X['doors'])
X['persons']=enc.fit_transform(X['persons'])
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 10)
X_train
```

```python
from imblearn.over_sampling import SMOTE
smote=SMOTE()
```

```python
X_train_smote,y_train_smote=smote.fit_resample(X_train.astype('float'),y_train)
```

```python
from collections import Counter
print("Before SMOTE: ",Counter(y_train))
print("After SMOTE: ", Counter(y_train_smote))
```

```python
model = KNeighborsClassifier()
model.fit(X_train_smote, y_train_smote)
y_predict = model.predict(X_test)
```

```python
print(accuracy_score(y_test,y_predict))
pd.crosstab(y_test,y_predict)
```
